## 1. Use your previous pre-processed dataset, keep the variables as one-hot encoded, and develop a multiple linear regression model. How many regression coefficients does this model have?

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42,shuffle=True)

model.fit(X_train, y_train)

num_coefficients = len(model.coef_) + 1 # Add the intercept

print("Number of regression coefficients:", num_coefficients)
```

## 3. Now go back to the original sex, age, and generation variables in their original numerical form. How many line coefficients are there?

```python
filename = "master.csv"
df = pd.read_csv(filename)

df.rename(columns={"HDI for year":"HDI_for_year",
                  " gdp_for_year ($) ":"gdp_for_year"," gdp_per_capita ($) ":"gdp_per_capita",
                    "gdp_per_capita ($)":"gdp_per_capita"}, inplace=True)
df["gdp_for_year"] = df["gdp_for_year"].str.replace(",","").astype(np.int64)
df["age"] = df["age"].str.replace("5-14 years","05-14 years")

age_mode = df['age'].mode()[0]
df.fillna({'age': age_mode}, inplace=True)
most_common_country = df['country'].mode()[0]
most_recent_year = df['year'].max()
df
```

Convert into numerical variables

```python
df['sex'] = df['sex'].map({'male': 1, 'female': 0})
age_mapping = {
    '05-14 years': 10,  # Approximate midpoint
    '15-24 years': 20,
    '25-34 years': 30,
    '35-54 years': 45,
    '55-74 years': 65,
    '75+ years': 85
}
df['age'] = df['age'].map(age_mapping)

generation_mapping = {
    'Generation Z': 6,
    'Millenials': 5,
    'Generation X': 4,
    'Boomers': 3,
    'Silent': 2,
    'G.I. Generation': 1
}
df['generation'] = df['generation'].map(generation_mapping)

df=df.drop(columns=['HDI_for_year', 'suicides_no', 'country-year','gdp_per_capita','population'])
df
```

```python
df = pd.get_dummies(df, 'country',drop_first=True)

X = df.drop(columns=['suicides/100k pop'])
y = df['suicides/100k pop']


model = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42,shuffle=True)

model.fit(X_train, y_train)

num_coefficients = len(model.coef_) + 1 # Add the intercept

print("Number of regression coefficients:", num_coefficients)

```

The number of coefficients is 106.

```python
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (MAE):", mae)
```

## 7. Give one advantage when using regression (as opposed to classification with nominal features) in terms of independent variables.
The advantage is its flexibility in handling continuous data. While classification models typically deal with categorical outcomes, regression can find the linear or non-linear relationships between continuous independent variables and a dependent variable. It can quantitatively predict the result and the magnitude indicates the correlation. When there are some flexible changes in input data, such as "Generation Alpha"; new range of age, regression can still present relatively accurate result. However, classification may fail to predict. Also, the outcome for classification cannot suggest the strength of predictions.

## 8. Give one advantage when using regular numerical values rather than one-hot encoding for regression.
The advatage is that it can suggest the ordinal relationships and reduce the dimensionality. One-hot encoding separate the features into different variables but converting into numerical values preserve the ordinal relationships in the model such as age range and generation types. Also, it have fewer dimensionality considering the coefficients, leading to better performance.
